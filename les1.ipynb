### Линейное пространство. Основные понятия. Часть 1

####__1.__ Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

Мы можем выделить линейную комбинацию векторов относителено $f_{4}: f_{4}= f_{3}-f_{2}-f_{1}.$
$$x-e^{x}=x+1-1-e^{x}.$$
$$x-e^{x}=x-e^{x}.$$
Векторы $f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}$ -линейно завилимы

####__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

Можно раскрыть квадрат суммы $f_{4}(x)=(x+1)^{2}=x^{2}+2x+1$ и задать линейную комбинацию векторов $$f_{4}(x)=f_{3}(x)+2f_{2}(x)+0.5f_{1}(x)$$
Следовательно векторы линейно зависимы.

####__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

а) $$(0, a, b)+(0, c, d) = (0, a+c, b+d),$$ $$\alpha\cdot (0, a, b) = (0, \alpha a, \alpha b).$$
$$(a, 0, b)+(c, 0, d) = (a+c, 0, b+d),$$ $$\alpha\cdot (a, 0, b) = (\alpha a, 0, \alpha b).$$
$$(0, 0, a)+(0, 0, c) = (0, 0, a+c),$$ $$\alpha\cdot (0, 0, a) = (0, 0, \alpha a).$$
Вектора является подпространством линейного пространства <br>
б) $$\{\lambda u_{1}, \lambda u_{2}, \ldots, \lambda u_{n}\} + \{\xi u_{1}, \xi u_{2}, \ldots, \xi u_{n}\} = \{(\lambda+\xi) u_{1}, (\lambda+\xi) u_{2}, \ldots, (\lambda+\xi) u_{n}\} $$
$$\alpha\cdot \{\lambda u_{1}, \lambda u_{2}, \ldots, \lambda u_{n}\} = \{\alpha\cdot\lambda u_{1}, \alpha\cdot\lambda u_{2}, \ldots, \alpha\cdot\lambda u_{n}\}$$
Вектора является подпространством линейного пространства
